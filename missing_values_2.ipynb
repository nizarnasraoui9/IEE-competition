```python
indexc=dfc.meter_id.tolist()
indexi=dfi.meter_id.tolist()
```

```python
def generate_null_values(df):
    null_values_x=[]
    null_values_y=[]
    for i in range (0,3248,1):
        for j in range(2,3,1):
            if (math.isnan(df.iloc[i,j])):
                null_values_x.append(i)
                null_values_y.append(j)
    return [null_values_x,null_values_y]
```

```python
[null_values_x,null_values_y]=generate_null_values(dfc)
```

```python
def remplace_by_having_same_dt_nb_noc(dfi,dfc,null_values_x,null_values_y,indexc,indexi):
    for i in range (0,len(null_values_x)):
        replaced=False
        smart_meter=indexc[null_values_x[i]]
        if(smart_meter in indexi):
            
            index_in_consumption=indexc.index(smart_meter) #the index of the smart meter in consumption dataframe
            index_in_addinfo=indexi.index(smart_meter) #the index of the smart meter in addinfo dataframe
            
            dwelling_type=dfi.iloc[index_in_addinfo,1] # dwelling dype of the smart meter
            number_of_bedrooms=dfi.iloc[index_in_addinfo,3] # number_of_bedrooms of the smart meter
            number_of_occupants=dfi.iloc[index_in_addinfo,2] # number of the occupants of the smart meter
            # ascending loop to fetch the smart meter having same dwelling_type,number_of_bedrooms_and number_of_occupants
            # and replace by the value finded in the missing value time 
            for j in range(index_in_addinfo+1,len(indexi)):
                #comparing with other smart meters features
                if((dfi.iloc[j,1]==dwelling_type) and (dfi.iloc[j,2]==number_of_occupants) and (dfi.iloc[j,3]==number_of_bedrooms)):
                    index_of_j_in_consumption=indexc.index(dfi.iloc[j,0])
                    tmp=dfc.iloc[index_of_j_in_consumption,null_values_y[i]]
                    #checking if finded value is not null
                    if (not math.isnan(tmp)):
                        dfc.iloc[null_values_x[i],null_values_y[i]]=tmp
                        print("i= ",i,"j= ",null_values_y[i],"is replaced by i= ",j,"j= ",null_values_y[i])
                        replaced=True
                        break
            # if the ascending loop fails to fetch the value we try the same steps using a descendent loop
            if(not replaced):
                for j in range(index_in_addinfo-1,0,-1):
                    if((dfi.iloc[j,1]==dwelling_type) and (dfi.iloc[j,2]==number_of_occupants) and (dfi.iloc[j,3]==number_of_bedrooms)):
                        index_of_j_in_consumption=indexc.index(dfi.iloc[j,0])
                        tmp=dfc.iloc[index_of_j_in_consumption,null_values_y[i]]
                        if (not math.isnan(tmp)):
                            dfc.iloc[null_values_x[i],null_values_y[i]]=tmp
                            print("i= ",i,"j= ",null_values_y[i],"is replaced by i= ",j,"j= ",null_values_y[i])
                            break
            
```
