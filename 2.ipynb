```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import interp1d
```

```python
df=pd.read_csv('QualityCover-results.csv')
df_density=pd.read_csv('QualityCover-results.csv',index_col='Density')
df_nstress=pd.read_csv('QualityCover-results.csv',index_col='N Stress')
df_nstress=pd.read_csv('QualityCover-results.csv',index_col='M stress')
df_qc=pd.read_csv('QualityCover-results.csv',index_col='QualityCover')
df_gress=pd.read_csv('QualityCover-results.csv',index_col='Gress')
```

```python
df.head()
```

```python
print("Gress-QualityCover en fonction de Density:")
print("f(density)=Gress-QualityCover:")
print("Si Gress-QualityCover<0 alors: Gress est meilleur que QualityCover, sinon si Gress-QualityCover>0 alors QualityCover est meilleur que Gress.")
plt.figure(figsize=(20,10))
plt.plot(df['Density'].values.tolist(),df['Gress']-df['QualityCover'])
plt.xlabel("Density")
plt.ylabel("Gress-QualityCover")
```

```python
print("Gress-QualityCover en fonction de M stress:")
print("f(M stress)=Gress-QualityCover:")
print("Si Gress-QualityCover<0 alors: Gress est meilleur que QualityCover, sinon si Gress-QualityCover>0 alors QualityCover est meilleur que Gress.")
plt.figure(figsize=(20,10))
plt.scatter(df['M stress'],df['Gress']-df['QualityCover'])
plt.ylabel("Gress-QualityCover")
plt.xlabel("M stress")
plt.show()
```

```python
print("Gress-QualityCover en fonction de N stress:")
print("f(M stress)=Gress-QualityCover:")
print("Si Gress-QualityCover<0 alors: Gress est meilleur que QualityCover, sinon si Gress-QualityCover>0 alors QualityCover est meilleur que Gress.")
plt.figure(figsize=(20,10))
plt.scatter(df['N Stress'],df['Gress']-df['QualityCover'])
plt.ylabel("Gress-QualityCover")
plt.xlabel("N Stress")
plt.show()
```
