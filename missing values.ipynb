```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df=pd.read_csv('updatedConsumption.csv')
```

```python
first_season_start=df.columns.get_loc("2017-01-01 00:00:00")
first_season_end=df.columns.get_loc("2017-02-28 23:30:00")
second_season_start=df.columns.get_loc("2017-11-01 00:00:00")
second_season_end=df.columns.get_loc("2017-12-31 23:30:00")
third_season_start=df.columns.get_loc("2017-03-01 00:00:00")
third_season_end=df.columns.get_loc("2017-04-30 23:30:00")
forth_season_start=df.columns.get_loc("2017-09-01 00:00:00")
forth_season_end=df.columns.get_loc("2017-10-31 23:30:00")
fifth_season_start=df.columns.get_loc("2017-05-01 00:00:00")
fifth_season_end=df.columns.get_loc("2017-08-31 23:30:00")
#weekdays
first_season_first_saturday=df.columns.get_loc("2017-01-07 00:00:00")
first_season_first_sunday=df.columns.get_loc("2017-01-01 00:00:00")
#second_season
second_season_first_saturday=df.columns.get_loc("2017-11-04 00:00:00")
second_season_first_sunday=df.columns.get_loc("2017-11-05 00:00:00")
#third_season
third_season_first_saturday=df.columns.get_loc("2017-03-04 00:00:00")
third_season_first_sunday=df.columns.get_loc("2017-03-05 00:00:00")
#forth_season
forth_season_first_saturday=df.columns.get_loc("2017-09-02 00:00:00")
forth_season_first_sunday=df.columns.get_loc("2017-09-03 00:00:00")
#fifth_season
fifth_season_first_saturday=df.columns.get_loc("2017-05-06 00:00:00")
fifth_season_first_sunday=df.columns.get_loc("2017-05-07 00:00:00")
```

```python
def gerenate_weekdays(df,first_saturday,first_sunday,end):
    weekdays=[]
    for i in range (first_saturday,end,288): #i is the first column of the weekday, i+47 is the last one
        for k in range(i,i+47,1):
            weekdays.append(k)
    for i in range (first_sunday,end,288):
        for k in range(i,i+47,1):
            weekdays.append(k)
    return weekdays
```

```python
first_season_weekdays=gerenate_weekdays(df,first_season_first_saturday,first_season_first_sunday,first_season_end)
second_season_weekdays=gerenate_weekdays(df,second_season_first_saturday,second_season_first_sunday,second_season_end)
third_season_weekdays=gerenate_weekdays(df,third_season_first_saturday,third_season_first_sunday,third_season_end)
forth_season_weekdays=gerenate_weekdays(df,forth_season_first_saturday,forth_season_first_sunday,forth_season_end)
fifth_season_weekdays=gerenate_weekdays(df,fifth_season_first_saturday,fifth_season_first_sunday,fifth_season_end)
```

```python
def gerenate_normal_days(df,first,last,season_week_days):
    normal_days=[]
    for i in range(first,last,1):
        if (i not in season_week_days):
            for k in range(i,i+47,1):
                normal_days.append(k)
    return normal_days
```

```python
first_season_normal_days=gerenate_normal_days(df,first_season_start,first_season_end,first_season_weekdays)
second_season_normal_days=gerenate_normal_days(df,second_season_start,second_season_end,first_season_weekdays)
third_season_normal_days=gerenate_normal_days(df,third_season_start,third_season_end,third_season_weekdays)
forth_season_normal_days=gerenate_normal_days(df,forth_season_start,forth_season_end,forth_season_weekdays)
fifth_season_normal_days=gerenate_normal_days(df,fifth_season_start,fifth_season_end,fifth_season_weekdays)
```

```python
def generate_null_values(df,start,end,season_week_days):
    x_null_non_weekdays=[]
    y_null_non_weekdays=[]
    x_null_weekdays=[]
    y_null_weekdays=[]
    for i in range (0,74,1):
        for j in range(start,end,1):
            if (math.isnan(df.iloc[i,j])) and (j not in season_week_days):
                x_null_non_weekdays.append(i)
                y_null_non_weekdays.append(j)
            elif(math.isnan(df.iloc[i,j])) and (j in season_week_days):
                x_null_weekdays.append(i)
                y_null_weekdays.append(j)
    return [x_null_non_weekdays,y_null_non_weekdays,x_null_weekdays,y_null_weekdays]
```

```python
[first_season_missing_values_x_non_weekdays,first_season_missing_values_y_non_weekdays,first_season_missing_values_x_weekdays,first_season_missing_values_y_weekdays]=generate_null_values(df,first_season_start,first_season_end,first_season_weekdays)
[second_season_missing_values_x_non_weekdays,second_season_missing_values_y_non_weekdays,second_season_missing_values_x_weekdays,second_season_missing_values_y_weekdays]=generate_null_values(df,second_season_start,second_season_end,second_season_weekdays)
[third_season_missing_values_x_non_weekdays,third_season_missing_values_y_non_weekdays,third_season_missing_values_x_weekdays,third_season_missing_values_y_weekdays]=generate_null_values(df,third_season_start,third_season_end,third_season_weekdays)
[forth_season_missing_values_x_non_weekdays,forth_season_missing_values_y_non_weekdays,forth_season_missing_values_x_weekdays,forth_season_missing_values_y_weekdays]=generate_null_values(df,forth_season_start,forth_season_end,forth_season_weekdays)
[fifth_season_missing_values_x_non_weekdays,fifth_season_missing_values_y_non_weekdays,fifth_season_missing_values_x_weekdays,fifth_season_missing_values_y_weekdays]=generate_null_values(df,fifth_season_start,fifth_season_end,fifth_season_weekdays)
```
