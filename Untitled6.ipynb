```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('addInfo.csv')
```

```python
print("visualisation des donn√©es manquantes: ")
sns.heatmap(df.isnull(), cbar=False)
```

```python
dwelling_type=['semi_detached_house','detached_house','terraced_house','bungalow','flat']
heating_fuel=['gas','elec','other','lpg_oil']
hot_water_fuel=['gas','elec','other']
boiler_age=['new','old']
loft_insulation=['y','n']
wall_insulation=['y_cavity','n','y_internal','not_sure','n']
heating_temperature=['below_18','18_to_20','above_20','not_sure']
efficient_lighting_percentage=['0_to_25','25_to_50','50_to_75','75_to_100']
df['dwelling_type'].replace(dwelling_type, [1,2,3,4,5], inplace=True)
df['heating_fuel'].replace(heating_fuel, [1,2,3,4], inplace=True)
df['hot_water_fuel'].replace(hot_water_fuel, [1,2,3], inplace=True)
df['boiler_age'].replace(boiler_age, [1,2], inplace=True)
df['loft_insulation'].replace(loft_insulation, [1,2], inplace=True)
df['wall_insulation'].replace(wall_insulation, [1,2,3,4,5], inplace=True)
df['heating_temperature'].replace(heating_temperature, [1,2,3,4], inplace=True)
df['efficient_lighting_percentage'].replace(efficient_lighting_percentage, [1,2,3,4], inplace=True)

print("dwelling_type:")
for i in range (0,len(dwelling_type)):
    print(str(i+1)+" :" +str(dwelling_type[i]))
print("heating_fuel:")
for i in range (0,len(heating_fuel)):
    print(str(i+1)+" :" +str(heating_fuel[i]))
print("hot_water_fuel:")
for i in range (0,len(hot_water_fuel)):
    print(str(i+1)+" :" +str(hot_water_fuel[i]))
print("boiler_age:")   
for i in range (0,len(boiler_age)):
    print(str(i+1)+" :" +str(boiler_age[i]))
print("loft_insulation:")
for i in range (0,len(loft_insulation)):
    print(str(i+1)+" :" +str(loft_insulation[i]))
print("wall_insulation:")
for i in range (0,len(wall_insulation)):
    print(str(i+1)+" :" +str(wall_insulation[i]))
print("heating_temperature:")
for i in range (0,len(heating_temperature)):
    print(str(i+1)+" :" +str(heating_temperature[i]))
print("efficient_lighting_percentage:")
for i in range (0,len(efficient_lighting_percentage)):
    print(str(i+1)+" :" +str(efficient_lighting_percentage[i]))
```

```python
df.hist(figsize=(16, 16),bins=16,linewidth='1',edgecolor='k',grid=False)
plt.show()
```

```python
for i in range (1,len(dwelling_type)):
    print("histogrammes ayant le dwelling_type= " + str(i))
    ht=(df['dwelling_type']==i)
    df[ht].hist(figsize=(16, 13),bins=16,linewidth='1',edgecolor='k',grid=False)
    plt.show()
```

```python
for i in range (1,5):
    print("histogrammes ayant le num_occupants= " + str(i))
    ht=(df.num_occupants==i)
    df[ht].hist(figsize=(16, 13),bins=16,linewidth='1',edgecolor='k',grid=False)
    plt.show()
```

```python
for i in range (1,len(heating_fuel)):
    print("histogrammes ayant le heating_fuel= " + str(i))
    ht=(df['dwelling_type']==i+1)
    df[ht].hist(figsize=(16, 13),bins=16,linewidth='1',edgecolor='k',grid=False)
    plt.show()
```

```python
print("histogramme ayant le boiler_age= new")
ht=(df.boiler_age==1)
df[ht].hist(figsize=(14, 9),bins=16,linewidth='1',edgecolor='k',grid=False)
plt.show()
print("histogramme ayant le boiler_age= old")
ht=(df.boiler_age==2)
df[ht].hist(figsize=(14, 9),bins=16,linewidth='1',edgecolor='k',grid=False)
plt.show()
print("remarque: il n'existe pas de maisons ou le boiler_age est old ayant 3 tablettes")
```

```python
for i in range (0,len(hot_water_fuel)):
    print("histogramme ayant le hot_water_fuel= " + hot_water_fuel[i])
    ht=(df.hot_water_fuel==i+1)
    df[ht].hist(figsize=(14, 9),bins=16,linewidth='1',edgecolor='k',grid=False)
    plt.show()
```
